```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline 

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15,7)
```

```python
arr = np.load("AA00400AA.npy",allow_pickle=True)
```

```python
np.shape(arr[0][0])
```

```python
k = 15
fig, axes = plt.subplots(k,1, figsize=(15,15), sharex=True, sharey=True)

for i in range(k):
    obs = arr[i]
    time = obs[:,0]
    current = obs[:,1]

    g= sns.lineplot(x=time,y=current, ax=axes[i])
    g.set_xlabel("Time [ms]")
```

```python
k = 5
fig, axes = plt.subplots(1,1, figsize=(15,15), sharex=True,)

for i in range(k):
    obs = arr[i]
    time = obs[:,0]
    current = obs[:,1]

    g= sns.lineplot(x=time,y=current)
    g.set_xlabel("Time [ms]")
```
